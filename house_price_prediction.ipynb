```python
import numpy as np
```

```python
import pandas as pd
```

```python
data=pd.read_csv("housing.csv (1).zip")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data["total_bedrrom_ratio"]=data["total_bedrooms"]/data["total_rooms"]
```

```python
data
```

```python
data=data.drop(["total_rooms","total_bedrooms"],axis=1)
```

```python
data
```

```python
inputs=data.drop(["median_house_value"],axis=1)
target=data["median_house_value"]
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
inputs["ocean_proximity"]=le.fit_transform(inputs["ocean_proximity"])
```

```python
inputs
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_inputs=sc.fit_transform(inputs)
scaled_inputs=pd.DataFrame(scaled_inputs,columns=inputs.columns)
```

```python
scaled_inputs
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,target,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfg=RandomForestRegressor(random_state=42)
rfg.fit(x_train,y_train)
```

```python
rfg.score(x_test,y_test)
```

```python
rfg.score(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100],
    'max_depth': [10, None],
    'min_samples_split': [2, 5],
}

grid_search=GridSearchCV(estimator=rfg,param_grid=param_grid,scoring="r2",cv=2,n_jobs=1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.score(x_test,y_test)
```
